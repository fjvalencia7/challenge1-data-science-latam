# 4. Productos más y menos vendidos

```python
product_sales_by_store = all_stores.groupby(['store', 'Producto']).agg(
    Cantidad_Vendida=('Producto', 'count'),
    Total_Ventas=('Total_Sales', 'sum')
).reset_index()

most_sold_product_with_sales = product_sales_by_store.loc[product_sales_by_store.groupby('store')['Cantidad_Vendida'].idxmax()].reset_index(drop=True)
display(most_sold_product_with_sales)
```

```python
least_sold_product_with_sales = product_sales_by_store.loc[product_sales_by_store.groupby('store')['Cantidad_Vendida'].idxmin()].reset_index(drop=True)
display(least_sold_product_with_sales)
```
